```python
######################### matplotlib (graphiques important) ###################
# %matplotlib : ##visulaiser le graphe de toutes les cotes
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D ### les 3D
import numpy as np

iris=load_iris()
x=iris.data 
y=iris.target 
print("nombre de classes est ",np.unique(y).size) ##np.unique caclule le nombre repetition et leur size
plt.scatter(x[:, 0],x[:,1],c=y,alpha=2,s=100) ##afficher toutes la data selon le premier target et le deuxieme target 


ax1=plt.axes(projection='3d')
ax1.scatter(x[:,0],x[:,1],x[:,2],c=y) ###3d 

fig2=plt.figure()
plt.hist(x[:,0]) ##hsitogramme pour la premiere colonne (1 er varaiable )
#plt.hist(x[:,1])
plt.hist2d(x[:,1],x[:,2]) ##histogramme 2d pour les eux premiere colonne (2 premiere varaibles )
plt.xlabel("longeure sepale")
plt.ylabel("largeure sepale ")
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from mpl_toolkits.mplot3d import Axes3D ### les 3D
import numpy as np

x=np.linspace(0,10,100)
y=np.linspace(0,10,100)
X,Y=np.meshgrid(x,y)
fonction=np.sin(X)
print("la taille de la grille est",fonction.shape)
ax2=plt.axes(projection="3d")
ax2.plot_surface(X,Y,fonction,cmap='plasma')
plt.show()

```
