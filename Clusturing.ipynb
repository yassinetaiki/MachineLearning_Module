```python
from sklearn.cluster import KMeans 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 
import numpy as np

iris=load_iris()
x=iris.data
y=iris.target

##pre_processing
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.8,random_state=0)
print(x.shape)

##modele
#miminimise inertie distance entre points xi du cluster et son center uj sum(min||(xi-mu)||²)
model=KMeans(n_clusters=3)
model.fit(xtrain)
print("prediction =>",model.predict(xtrain))
print("centres  =>",model.cluster_centers_)
print(" inertia =>",model.inertia_)

##elbow_methode(#detecter meilleures nombre de cluster)
cout=[]
for i in range(1,10,1):
    modele2=KMeans(n_clusters=i)
    modele2.fit(xtrain)
    cout.append(modele2.inertia_)


j=np.arange(1,10) ##verfier les coudes dans ce cas c'est  3 clusters
fig1=plt.figure()
plt.plot(j,cout,c='b') 
plt.xlabel('nbr_cluster')
plt.ylabel('inertia')
plt.title(' evolution inertia')

#visualisation
fig2=plt.figure()
plt.scatter(xtrain[:,0],xtrain[:,1],c=model.predict(xtrain))
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c='r')







```

```python
from  sklearn.cluster import AffinityPropagation
from sklearn.cluster import DBSCAN

##AffinityPropagation##
modele3=AffinityPropagation()
modele3.fit(xtrain)
print("prediction 2=>",modele3.predict(xtrain))
print(" center 2=> ",modele3.cluster_centers_)
##visualisation
fig3=plt.figure()
plt.scatter(xtrain[:,0],xtrain[:,1],c=model.predict(xtrain))
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],c='r')


##DBSCAN##
modele4=DBSCAN()
modele4.fit(xtrain)
print("")
print("prediction 3=>",modele4.labels_) ##ehantillons bruyant recoivent label -1
print(" center 2=> ",modele4.core_sample_indices_)
##visualisation
fig4=plt.figure()
plt.scatter(xtrain[:,0],xtrain[:,1],c=model.predict(xtrain))




```
