```python
###################### traitement de signal #####################
from scipy import signal
from scipy import fftpack

x3=np.linspace(0,20,100)
y1=x3+np.sin(x3)+np.random.randn(100)
plt.plot(x3,y1)
new_y=signal.detrend(y1) ##signal sans tendance lineaire 
plt.plot(x3,new_y,c='r')
plt.show()


fourier = fftpack.fft(y1) ##transformer de fourier  
frequence = fftpack.fftfreq(y1.size) ##frequence
fig1=plt.figure()
plt.plot(np.abs(frequence),np.abs(fourier)) ##valeur absolue pour eviter les valeures negative 

fourier[np.abs(fourier)<200]=0  ##enelever le bruit du signal 

signal_filtree=fftpack.ifft(fourier) ##on filtere le signal avec la transfrome de fouriere inverse 
fig3=plt.figure()
plt.plot(x3,y1,c='b')
plt.plot(x3,signal_filtree,c='r')
plt.show()

```
